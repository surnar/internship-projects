```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sample movies data
data={
    'movie_title':['movie a','movie b','movie c','movie d','movie e','movie f','movie g','movie h','movie i','movie j'],
    'genre':['action','romance','comedy','drama','action','comedy','romance','thriller','drama','action'],
    'release_year':[2023,2022,2023,2021,2022,2023,2020,2022,2021,2023],
    'box_office_collection':[15000000000,8000000000,12000000000,6000000000,20000000000,10000000000,9000000000,7000000000,5000000000,18000000000],
    'lead_actor':['actor x','actor y','actor z','actor x','actor y','actor z','actor x','actor y','actor z','actor x']
}
df=pd.DataFrame(data)
```

```python
# basic data exploration
df.head()
```

```python
df.info()
```

```python
# genre analysis
genre_counts=df['genre'].value_counts()
```

```python
genre_counts
```

```python
# box office collection by genres
box_office_by_genre=df.groupby('genre')['box_office_collection'].sum()
```

```python
box_office_by_genre
```

```python
# lead actor analysis
lead_actor_box_office=df.groupby('lead_actor')['box_office_collection'].sum()
```

```python
lead_actor_box_office
```

```python
# bar chart of movie count by genre
plt.figure(figsize=(10,6))
sns.countplot(x='genre',data=df)
plt.title('number of movies by genre')
plt.xlabel('genre')
plt.ylabel('number of movies')
plt.show()
```

```python
# pie chart of box office collection by genre
plt.figure(figsize=(8,8))
plt.pie(box_office_by_genre, labels=box_office_by_genre.index, autopct='%1.1f%%',startangle=140)
plt.title('box office collection by genre')
plt.show()
```
